
<h4 Style = "color:#80b0ab"> 2.3 Imputación simple de valores númericos </h4>

<p > No siempre se van a imputar todos los datos ya que se debe validar en cada caso dependiendo el contexto del análisis esto se consulta con el experto del tema, a validar si es posible suplantar estos datos.</p>

```python
promNormalizadas= dfAutos["normalized"].astype("float").mean(axis=0)
dfAutos["normalized"].replace(np.nan, promNormalizadas, inplace=True)
dfAutos.head(5)
#Esto lo que hizo fue reemplazar los datos faltantes con un promedio de los datos existentes, siempre con datos "float"
```

```python
promBore= dfAutos["bore"].astype("float").mean(axis=0)
dfAutos["bore"].replace(np.nan, promBore, inplace=True)
dfAutos.head(5)
```

```python
promStroke= dfAutos["stroke"].astype("float").mean(axis=0)
dfAutos["stroke"].replace(np.nan, promStroke, inplace=True)
dfAutos.head(5)
```

```python
promHorsepower= dfAutos["horsepower"].astype("float").mean(axis=0)
dfAutos["horsepower"].replace(np.nan, promStroke, inplace=True)
dfAutos.head(5)
```

```python
promPeak= dfAutos["peak"].astype("float").mean(axis=0)
dfAutos["peak"].replace(np.nan, promStroke, inplace=True)
dfAutos.head(5)
```


<h4 Style = "color:#80b0ab"> 2.4 Imputación simple de valores Categoricos </h4>


<p> Cuando son imputaciones de caracteres o tipo object, se realiza diferentes se hace por el valor más frecuente</p>

```python
dfAutos["doors"].value_counts().idxmax()
#dfAutos["doors"]: Accede a la columna "doors" del DataFrame dfAutos.
#value_counts(): Calcula el recuento de ocurrencias de cada valor único en la columna 
#.idxmax(): Identifica el valor que tiene la máxima frecuencia. Devuelve el valor único que aparece con mayor frecuencia
```

```python
dfAutos["doors"].replace(np.nan, "four", inplace=True)
dfAutos.head()
```

<h4 Style = "color:#80b0ab"> 4.1 Configuración para los gráficos </h4>

```python
#Remueve las lineas de los axis.
mpl.rcParams["axes.spines.bottom"] = True
mpl.rcParams["axes.spines.left"] = False
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
gf = dfAutos.groupby(['make']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('symboling', ascending=False)
gf
```

```python
fig, ax = plt.subplots(1,1, figsize =(15,6.9), dpi=100)

 

eje_x = gf.index.tolist()
data = gf["symboling"].tolist()

 

bar_colors = ['#69d2e7','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5',
             '#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#e5e5e5','#fa6900']

 

ax.barh(eje_x, data, color=bar_colors)

 

#--------------------------------------------------------------------------------------------#
#-----------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EFES DEL GRAFICO----------------#
#--------------------------------------------------------------------------------------------#
#FORMATO DEL TITULO

 

title = "Grafico 1: Vehiculos en Inventario"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             )

 

#FORMATO DEL SUBTITULO
subTitle = "Cantidad de vehiculos agrupados segun fabricante, en Costa Rica, periodo 2023"
ax.set_title(subTitle,
             horizontalalignment = "left",
             x = 0.001,
             y = 0.80,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

 

 

#FORMATO AL EJE (X/Y)
etiqueta = "cantidad de carros (ABS)"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

 

ax.xaxis.set_label_coords(x=0, y=-0.07, transform = ax.transAxes)

 

#FORMATO TAMANO LETRA AMBOS EJES
ax.tick_params(axis='both', labelsize = 10)

 

#CONFIGURA LA UBICACION EL TITULO, SUBTITULO Y GRAFICO
plt.subplots_adjust(top = 0.80)

 

plt.show()
```
